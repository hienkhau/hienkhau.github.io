Q.4) How do the Average Views, Likes, and Comments are compared between different Album Types?

```python
dt = df.groupby("Album_type")[["Likes","Views","Comments"]].mean()
dt
```

```python
dt.reset_index(inplace=True)
```

```python
dt.columns
```

```python
dt_melted = pd.melt(dt, id_vars = ['Album_type'], var_name = 'Attribute' , value_name = 'Total')
dt_melted
```

```python
plt.figure(figsize = (8,5))
sns.barplot(x = "Album_type", y = "Total", hue = "Attribute", data = dt_melted);
```
